```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Load the data you receive into a Pandas DataFrame

```python
df= pd.read_csv('menu.csv')
df.head()
```

### Analyze dataset

## Note1:
#### we can see by representing this graph that the quantity of protien Be as high as possible in lunches, as little as    possible in snaks nand moderate in breakfasts , so high protine by McDonald's in Chicken & Fish and thisis the normal , and   the lowest im Beverages

```python
sns.relplot(x='Category', y='Protein', data=df, kind='line',aspect=2.5)

```

# ______________________________________________________________

## Note2:
#### we can see by representing this graph that the Chiken and fish hada high number of  Calories ,and coffe and tea had the least nymber of calories so dink tea gyes  

```python
sns.barplot(x='Category', y='Calories', data=df,ci=None, palette="Blues_d")
sns.set(rc={'figure.figsize':(30,20)})
```

## Note3:
#### by groping the total fat and sum the result we can see the quantity od suger between (0-200) , have zeroo Trans fat 

```python
fat=df.groupby('Total Fat (% Daily Value)',as_index=False).sum()
fat
```

```python
sns.relplot(x='Sugars', y='Trans Fat', data=fat,palette="Blues_d")
```
